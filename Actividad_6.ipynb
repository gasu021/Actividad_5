# **Instrucciones**

Escribe un programa en Python que realice lo siguiente:

1. Codifica una funci√≥n que reciba una lista como argumento y que genere un diccionario como salida.
2. El diccionario generado debe tener las siguientes llaves (claves): nombre, apellido y edad.
3. Genera la lista de entrada con valores ficticios de posibles personas reales.
4. Ejecuta el programa tres veces con valores diferentes y almacena todos los resultados en la estructura de datos m√°s adecuada para la situaci√≥n planteada.
5. Escribe el diccionario del ejercicio anterior en un archivo CSV o JSON.
6. Usando alguna de las funciones de lectura de Python, despliega el contenido de lo anterior.
7. Por medio de la clase persona, modela la informaci√≥n del ejercicio anterior y crea una instancia que contenga tu nombre, apellido y edad como atributos.


```python
# ---SISTEMA DE DISTRIBUCI√ìN DE GATOS ---

# 1. Funci√≥n: lista como argumento y diccionario como salida.
# 2. Dicconario: nombre, apellido y edad.
def michi_sistema_distribucion(datos_lista):
    """Funci√≥n que act√∫a como el motor del Sistema de Distribuci√≥n.
    Transforma una lista de humano detectado en un expediente (diccionario)."""
    expediente_humano = {
        "nombre": datos_lista[0],
        "apellido": datos_lista[1],
        "edad": datos_lista[2]
    }
    return expediente_humano
```

```python

# 3. Lista de entrada con valores ficticios.
humanos_candidatos = [
    ["Susana", "Alfaro", 62],
    ["Roberto", "G√≥mez", 45],
    ["Luc√≠a", "Fern√°ndez", 28]
]
# 4. Ejecutar 3 veces
base_datos_distribucion = []

for candidato in humanos_candidatos:
    expediente = michi_sistema_distribucion(candidato)
    base_datos_distribucion.append(expediente)

print("üêæ [SISTEMA DE DISTRIBUCI√ìN DE GATOS]: Procesamiento completado.")
print(base_datos_distribucion)

```

```python
# 5. Escribir el diccionario en un archivo JSON
import json

archivo_sistema = "expediente_distribucion.json"

with open(archivo_sistema, "w") as michi_file:
    json.dump(base_datos_distribucion, michi_file, indent=4)
    print(f"\nüìÇ Archivo '{archivo_sistema}' generado.")

```

```python
# 6. Desplegar el contenido y mis datos
print("\nüîé Desplegando registros del Sistema de Distribuci√≥n:")
with open(archivo_sistema, "r") as lectura:
    datos_finales = json.load(lectura)
    for registro in datos_finales:
        print(f"Humano asignado: {registro['nombre']} {registro['apellido']}, Edad: {registro['edad']}")
```

```python
# 7. Clase persona
class Persona:
    def __init__(self, nombre, apellido, edad):
        self.nombre = nombre
        self.apellido = apellido
        self.edad = edad

# Creamos la instancia con tus datos reales
# Gabriela Escobar, 39 a√±os - Administradora del Sistema
michi_lider = Persona("Gabriela", "Escobar", 39)

print(f"\n‚ú® 'Persona' asignada ‚ú®")
print(f"Nombre: {michi_lider.nombre} {michi_lider.apellido}")
print(f"Edad: {michi_lider.edad} a√±os")
```
