# **Parte 1**

Utilizando la herramienta **Google Colab** programa lo siguiente:

* Escribe una lista que contenga un mínimo de cinco elementos.
* Agrega un nuevo elemento en el índice 2.
* Agrega un elemento al final de la lista.


---


* Escribe una cadena de caracteres que contenga tres palabras.
* Cuenta los caracteres de la cadena generada.
* Guarda la cadena resultante toda en mayúsculas en una nueva variable.


---


* Escribe un diccionario que contenga al menos cuatro elementos y cuyos índices coincidan con las primeras letras del alfabeto.
* Agrega un nuevo elemento al diccionario.
* Elimina el elemento del índice “c”.
* Cuenta los elementos que componen el diccionario después de completadas las operaciones anteriores.


```python
# --- 1. LISTAS ---
# Variable: lista_gatos
lista_gatos = ["Siamés", "Persa", "Maine Coon", "Bengalí", "Esfinge"]
print(f"Lista original: {lista_gatos}")

lista_gatos.insert(2, "Ragdoll")
lista_gatos.append("Siberiano")
print(f"Lista modificada: {lista_gatos}")


# --- 2. CADENAS DE CARACTERES ---
# Variable: cadena_gatos
cadena_gatos = "Gatos son geniales"

cantidad_caracteres = len(cadena_gatos)
print(f"\nCaracteres en cadena: {cantidad_caracteres} caracteres.")

cadena_mayusculas = cadena_gatos.upper()
print(f"Cadena en mayúsculas: {cadena_mayusculas}")


# --- 3. DICCIONARIOS ---
# Variable: diccionario_gatos
diccionario_gatos = {
    "a": "Cálico",
    "b": "Carey",
    "c": "Tabby",
    "d": "Tuxedo"
}
print(f"\nDiccionario original: {diccionario_gatos}")

diccionario_gatos["e"] = "Punto de color"
del diccionario_gatos["c"]

total_elementos = len(diccionario_gatos)
print(f"Diccionario final: {diccionario_gatos}")
print(f"El diccionario ahora tiene {total_elementos} elementos.")
```

**Conclusiones**

El uso de estas tres estructuras fundamentales permite gestionar la información de manera eficiente según las necesidades del programa:

1. **Listas (lista_gatos):** Son colecciones ordenadas y mutables. Son ideales cuando el orden de los elementos es importante y cuando necesitamos realizar operaciones constantes de inserción o eliminación de datos.

2. **Cadenas de caracteres (cadena_gatos):** Aunque funcionan como secuencias de caracteres, su manejo permite el procesamiento y transformación de texto, lo cual es vital para la interfaz con el usuario y el análisis de datos textuales.

3. **Diccionarios (diccionario_gatos):** Su estructura de pares clave-valor permite organizar datos de forma semántica. Son extremadamente eficientes para búsquedas rápidas, ya que permiten acceder a la información mediante etiquetas únicas en lugar de posiciones numéricas.

# **Parte 2**

Una sala de cine cobra diferentes precios de entrada según la edad de la persona. Si una persona es menor de 3 años, la entrada es gratuita; si tienen entre 3 y 12, el boleto es de $10; y si son mayores de 12 años, el boleto cuesta $ 15. Escribe un bucle en el que se solicite a los usuarios su edad y despliegue el costo de la entrada.

```python
# --- Parte 2: MICHICINE ---
# Inicializamos la variable
entrada_edad = ""

print("--- Bienvenido a la taquilla de MichiCine ---")
print("(Escribe 'salir' para terminar el programa)\n")

while entrada_edad.lower() != 'salir':
    entrada_edad = input("Por favor, ingresa tu edad (o escribe 'salir'): ")

    # Verificamos de nuevo si es 'salir' para no procesar el código de abajo
    if entrada_edad.lower() != 'salir':
        try:
            # Convertimos la entrada a número entero
            edad = int(entrada_edad)

            # Precios de MichiCine
            if edad < 3:
                resultado = "La entrada es gratuita."
            elif 3 <= edad <= 12:
                resultado = "El boleto cuesta $10."
            else:
                resultado = "El boleto cuesta $15."

            print(f"Resultado MichiCine: {resultado}\n")

        except ValueError:
            print("Error: Por favor, escribe una edad válida (numero entero) o la palabra 'salir'.\n")

print("\n¡Miauchas gracias por visitarnos! Que disfrutes la función.")
```
